```python
#Завдання 1
suicide = pandas.read_csv("/home/anatoliy/Downloads/master.csv")
print(suicide.columns)
print("Raws,Columns:")
print(suicide.shape)
suicide = suicide.dropna() 
suicide.isnull().sum()
```

```python
suicide_only_numbers  = suicide.drop(columns=['sex', 'age', 'suicides_no', 'country-year','population', 'generation', ' gdp_for_year ($) ', 'gdp_per_capita ($)'], axis=1).set_index('country') #Getting numbers
X = suicide_only_numbers.drop(['year'], axis=1)
Y = suicide_only_numbers['year']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.30)
print(X_train.shape)
print(Y_train.shape)
```

```python
model = Sequential()
model.add(layers.Dense(32, activation="relu"))
model.add(layers.Dense(32, activation="relu"))
model.add(Dense(1, activation='sigmoid'))
model.add(layers.Flatten())
model.build(input_shape=(None, 2)) 
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, Y_train, epochs=100, validation_split=0.2)
```

```python
Y_prediction = model.predict(X_test) > 0.5
print(confusion_matrix(Y_test, Y_prediction))
print(classification_report(Y_test, Y_prediction))
#Результат майже такий самий
```
